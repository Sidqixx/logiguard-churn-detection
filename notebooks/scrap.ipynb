```python
from google_play_scraper import Sort, reviews
import pandas as pd

target_app = 'com.msd.JTClient' 

result, _ = reviews(
    target_app,
    lang='id', 
    country='id', 
    sort=Sort.NEWEST, 
    count=3000 
)

df = pd.DataFrame(result)

if df.empty:
    print("No data obtained")
else:
    print(df[['userName', 'score', 'content']].head())

df.to_csv('jnt_reviews_raw.csv', index=False)
print(f"Obtained {len(df)} data review.")
```
