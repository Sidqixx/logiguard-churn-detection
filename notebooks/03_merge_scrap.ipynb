```python
import pandas as pd

df1 = pd.read_csv('../data/raw/jnt_reviews_raw.csv')
df2 = pd.read_csv('../data/raw/jnt_reviews_raw_2.csv')

df_all = pd.concat([df1, df2], ignore_index=True)
df_unique = df_all.drop_duplicates(subset=['userName', 'content'])

print(f"Total data unik setelah digabung: {len(df_unique)}")
print("\nDistribusi Rating Akhir:")
print(df_unique['score'].value_counts())

df_all.to_csv('../data/processed/jnt_reviews_final.csv', index=False)
```
