```python
import pandas as pd
import re

df = pd.read_csv('../data/jnt_reviews_raw.csv')

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'http\s+|www\s+|https\s+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = " ".join(text.split())
    return text

norm_map = {
    "gk": "tidak", "gak": "tidak", "nggak": "tidak", "ga": "tidak", "tdk": "tidak",
    "yg": "yang", "utk": "untuk", "dgn": "dengan", "udh": "sudah", "udah": "sudah",
    "bgt": "banget", "sdh": "sudah", "aja": "saja", "jd": "jadi", "kalo": "kalau",
    "klo": "kalau", "tp": "tapi", "pake": "pakai", "pke": "pakai", "blm": "belum",
    "tau": "tahu", "bangettt": "banget", "si": "saja", "dah": "sudah","d": "di"
}

def formalize_text(text):
    words = text.split()
    res = [norm_map.get(w, w) for w in words]
    return " ".join(res)

df['content_clean'] = df['content'].apply(clean_text)
df['content_clean'] = df['content_clean'].apply(formalize_text)

df = df[df['content_clean'] != ""]

df.to_csv('../data/jnt_reviews_cleaned.csv', index=False)

print("\nCheck comparison:")
print(df[['content', 'content_clean']].head(10))
```
