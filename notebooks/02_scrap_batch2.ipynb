```python
from google_play_scraper import Sort, reviews
import pandas as pd
import os

target_app = 'com.msd.JTClient'
file_name = '../data/raw/jnt_reviews_raw_2.csv'

res, _ = reviews(
    target_app, 
    lang='id', 
    country='id', 
    sort=Sort.NEWEST, 
    count=1000
)

df_new = pd.DataFrame(res)

df_pos = df_new[df_new['score'] >= 4]

if os.path.exists(file_name):
    df_existing = pd.read_csv(file_name)
    df_final = pd.concat([df_existing, df_pos]).drop_duplicates(subset=['userName', 'content', 'at'])
    df_final.to_csv(file_name, index=False)
    print(f"Total data rating positif terkumpul: {len(df_final)}")
else:
    # Kalau belum ada, kita buat baru
    df_pos.to_csv(file_name, index=False)
    print(f"Data rating positif awal terkumpul: {len(df_pos)}")
```
