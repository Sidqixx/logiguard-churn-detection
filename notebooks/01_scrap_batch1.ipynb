```python
from google_play_scraper import Sort, reviews
import pandas as pd
import time

target_app = 'com.msd.JTClient' 

res_relevant, _ = reviews(
    target_app, lang='id', country='id', 
    sort=Sort.MOST_RELEVANT, count=1500
)

res_newest, _ = reviews(
    target_app, lang='id', country='id', 
    sort=Sort.NEWEST, count=1500
)

combined_results = res_relevant + res_newest
df = pd.DataFrame(combined_results)

df = df.drop_duplicates(subset=['userName', 'content', 'at'])

print(f"Total data setelah drop duplikat: {len(df)}")

print("\nDistribusi Rating Saat Ini:")
print(df['score'].value_counts())

df.to_csv('../data/raw/jnt_reviews_raw.csv', index=False)
```
