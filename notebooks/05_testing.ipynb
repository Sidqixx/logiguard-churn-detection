```python
import pandas as pd
from transformers import pipeline
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

df_eval = pd.read_csv('../data/jnt_final_labeled.csv')
nlp = pipeline("sentiment-analysis", model="mdhugol/indonesia-bert-sentiment-classification")
```

```python
def get_prediction(text):
    res = nlp(str(text)[:512])[0]
    if res['label'] == 'LABEL_2':
        return 1 # Predicted Churn
    else:
        return 0 # Predicted Aman

df_eval['prediction'] = df_eval['content'].apply(get_prediction)

print("\n--- EVALUATION INDOBERT ---")
print(classification_report(df_eval['label'], df_eval['prediction'], target_names=['Aman', 'Churn']))

```

```python
cm = confusion_matrix(df_eval['label'], df_eval['prediction'])
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Aman', 'Churn'], yticklabels=['Aman', 'Churn'])
plt.xlabel('Prediksi AI')
plt.ylabel('Kenyataan (Label)')
plt.title('Confusion Matrix: LogiGuard V2')
plt.show()
```
